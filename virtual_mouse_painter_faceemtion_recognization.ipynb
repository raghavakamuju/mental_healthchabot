```python
# import numpy as np
# import pandas as pd
# import tensorflow
# from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow import keras

# model = keras.Sequential([
#     keras.layers.Conv2D(128, input_shape=(150, 150, 3), kernel_size=(3, 3), activation='relu'),
#     keras.layers.MaxPool2D((3, 3)),
#     keras.layers.Conv2D(128, (3, 3), activation='relu'),
#     keras.layers.MaxPool2D((3, 3)),
#     keras.layers.Conv2D(64, (3, 3), activation='relu'),
#     keras.layers.MaxPool2D((3, 3)),
#     keras.layers.Flatten(),
#     keras.layers.Dense(512, activation='relu'),
#     keras.layers.Dropout(0.5),
#     keras.layers.Dense(256, activation='relu'),
#     keras.layers.Dropout(0.2),
#     keras.layers.Dense(128, activation='relu'),
#     keras.layers.Dense(7, activation='softmax')
# ])

# model.compile(
#     optimizer='adam',
#     metrics=['accuracy'],
#     loss='categorical_crossentropy'
# )

# train_datagen = ImageDataGenerator(
#     rescale=1./255,
#     shear_range=0.2,
#     zoom_range=0.2,
#     horizontal_flip=True
# )

# training_samples = train_datagen.flow_from_directory(
#     "C:/Users/ragha/OneDrive/Desktop/mental_health_chatbot/train",
#     target_size=(150, 150),
#     batch_size=32,
#     class_mode='categorical'
# )

# model.fit(training_samples,epochs=5)
# import pickle
# with open('emotion_model.pkl','wb') as f:
#     pickle.dump(model,f)

import cv2
import numpy as np
import pickle
with open('emotion_model.pkl','rb') as f:
    model=pickle.load(f)
import cv2
import numpy as np

canvas = np.ones((500, 500, 3), dtype=np.uint8) * 255
color = (0, 0, 0)
drawing = False

def draw_circle(event, x, y, flags, param):
    global drawing

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        cv2.circle(canvas, (x, y), 6, (0, 0, 0), 5, -1)
    elif event == cv2.EVENT_MOUSEMOVE and drawing:
        cv2.circle(canvas, (x, y), 5, color, -1)
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        print('Pen lifted')

cv2.namedWindow('Color Painter')
cv2.setMouseCallback('Color Painter', draw_circle)

while True:
    cv2.imshow('Color Painter', canvas)
    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):  
        break
    elif key == ord('s'): 
        cv2.imwrite('output.jpg', canvas)

cv2.destroyAllWindows()   
img = cv2.imread('output.jpg', 1)
img = cv2.resize(img, (150, 150))

img = img / 255.0  
 
img = np.expand_dims(img, axis=0)  

prediction = model.predict(img)

print("Predictions:", prediction)

for i, pred in enumerate(prediction[0]):
    print(f'For class {i}: {pred}')
```
